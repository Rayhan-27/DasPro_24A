<h1>Data Sampel Kepuasaan Customer E-Commerce</h1>

```python
import numpy as np
import pandas as pd

n = 1000

id1 = np.random.randint(1, 1001, size = n)
customerId = ["ID" + str(i).zfill(4) for i in id1] #fungsi dari zfill(4) adalah untuk menambahkannol di depan supaya panjangnya sama yaitu 4

usia = np.random.randint(18,61,size = n) #maksimal 61 agar hasilnya maks sebenarnya 60

gender = np.random.choice(['Pria','Wanita'], size = n)

id2 = np.random.randint(10, 50500, size = n)
buyId = ["SPXID" + str(b).zfill(5) for b in id2]

produk = np.random.choice(['Electronics','Fashion','Food','Home','Sport'],size = n)

trsksi = np.random.randint(10,1000,size = n)
transaksi = ["Rp " + str(t) + ".000,-" for t in trsksi]

hari = np.random.randint(1, 31, size = n)
bulan = np.random.randint(1, 12, size = n)
tahun = np.random.choice(['2023','2024','2025'], size = n)
waktuPesan = [f"{str(h).zfill(2)}-{str(b).zfill(2)}-{t}" for h, b, t in zip(hari, bulan, tahun)]

jenisPaket = np.random.choice(['Express','Reguler','Next Day','Hemat'], size = n)

pembayaran = np.random.choice(['Cash','COD','Transfer','Ngutang'], size = n)

feedback = np.random.choice(['★★★★★','★★★★','★★★','★★','★'],size = n) #bila tidak memakai size = n_data, maka tidak akan random


df = pd.DataFrame({
    'Customer ID' : customerId,
    'Usia' : usia,
    'Gender' : gender,
    'ID Pembelian' : buyId,
    'Jenis Produk' : produk,
    'Transaksi' : transaksi,
    'Waktu Pembelian' : waktuPesan,
    'Jenis Pengiriman' : jenisPaket,
    'Metode Pembayaran' : pembayaran,
    'Feedback' : feedback
})

print(df)
print("")
print("-----------------------------------------------------------------------------")
print("")
print(df.info())
print("")
print("-----------------------------------------------------------------------------")
print("")
print(df.head())
df.to_csv('KepuasanECommerce.csv')
```
